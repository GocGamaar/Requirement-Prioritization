```python
from matplotlib import pyplot as plt
import numpy as np
import time


from GA import GeneticAlgorithm
```

```python
iterationsArray = [10, 25, 50, 100, 250]
```

```python
def obj_function(genome):
    x1, x2, x3, x4, x5, x6 = genome
    
    if x5==x6:
        x6=0
    if x4==x5 or x4==x6:
        x5 = x6 = 0
    if x4>x5 and x4>x6:
        x5 = x6 = 0
    if x5>x6 and x5>x4:
        x4 = x6 = 0
    if x6>x4 and x6>x5:
        x4 = x5 = 0
        
    if x3==0:
        return 0
    else:
        return x1 + (10-(x2/x3)) + (x4*9) + (x5*5) + (x6*1)
```

```python
bestFitArray = []; gaExeArray = [];
for i in range(iterationsArray[0]):
    ga_start = time.time()
    ga = GeneticAlgorithm(
        obj_function, 
        pop_size=100, 
        genome_length=6,
        lb=[0, 1, 1, 0, 0, 0 ],
        ub=[35, 8, 9, 1, 1, 1],
        timeplotbool = False
    )
    ga.generate_binary_population()
    ga.number_of_pairs = 4
    ga.selective_pressure = 1.3
    ga.mutation_rate = 0.1
    ga.run(25)
    best_genome, best_fitness = ga.get_best_genome()
    gaExeArray.append( time.time() - ga_start )
    bestFitArray.append(best_fitness)

print("Average best fitness for ",iterationsArray[0]," iterations: ", sum(bestFitArray)/(iterationsArray[0]))
print("Average GA exe time for ",iterationsArray[0]," iterations: ", sum(gaExeArray)/(iterationsArray[0]))
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True)
fig.set_size_inches(18, 10)
fig.suptitle(str(iterationsArray[0])+" iterations")
ax1.set_title("Best fitness values")
ax1.plot(np.arange(iterationsArray[0]), bestFitArray, linewidth=4.0, color="#FFC300", marker="o", mfc="#000000")

ax2.set_title("Execution time")
ax2.plot(np.arange(iterationsArray[0]), gaExeArray, linewidth=4.0, color="#FA8072", marker="o", mfc="#000000")

fig.show()
```
