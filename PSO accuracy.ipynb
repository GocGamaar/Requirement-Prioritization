```python
from matplotlib import pyplot as plt
import numpy as np
import time


from PSO import PSO
```

```python
iterationsArray = [10, 25, 50, 100, 250]
```

```python
def obj_function(genome):
    x1, x2, x3, x4, x5, x6 = genome
    
    if x5==x6:
        x6=0
    if x4==x5 or x4==x6:
        x5 = x6 = 0
    if x4>x5 and x4>x6:
        x5 = x6 = 0
    if x5>x6 and x5>x4:
        x4 = x6 = 0
    if x6>x4 and x6>x5:
        x4 = x5 = 0
        
    if x3==0:
        return 0
    else:
        return x1 + (10-(x2/x3)) + (x4*9) + (x5*5) + (x6*1)
```

```python
gBestArray = []; psoExeArray = [];
for i in range(iterationsArray[0]):
    pso_start = time.time()
    instance = PSO(
        func=obj_function,
        dim=6,
        lb=[0, 1, 1, 0, 0, 0 ],
        ub=[35, 8, 9, 1, 1, 1],
        w=0.6, c1=0.7, c2=0.7,
        timeplotbool = False
    )

    pso_result = instance.run(max_iter=25)

    psoExeArray.append(time.time()-pso_start)
    gBestArray.append(pso_result.gbest_y)

print("Average PSO gBest for ",iterationsArray[0]," iterations: ", sum(gBestArray)/(iterationsArray[0]))
print("Average PSO exe time for ",iterationsArray[0]," iterations: ", sum(psoExeArray)/(iterationsArray[0]))
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True)
fig.set_size_inches(18, 10)
fig.suptitle(str(iterationsArray[0])+" iterations")
ax1.set_title("gBest values")
ax1.plot(np.arange(iterationsArray[0]), gBestArray, linewidth=4.0, color="#FFC300", marker="o", mfc="#000000")

ax2.set_title("Execution time")
ax2.plot(np.arange(iterationsArray[0]), psoExeArray, linewidth=4.0, color="#FA8072", marker="o", mfc="#000000")

fig.show()
```
